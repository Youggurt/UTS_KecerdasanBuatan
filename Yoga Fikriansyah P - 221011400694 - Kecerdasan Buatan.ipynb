```python
# Nama : Yoga Fikriansyah Pangalo
# NIM : 2210114006944
# Data company & employee
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'Hong Kong'
    }
]

# Task 1 - Sort companies in reverse order
def get_sorted_companies():
    companies = [{'name': company['name']} for company in company_detail_list]
    sorted_companies = sorted(companies, key=lambda x: x['name'], reverse=True)
    print(sorted_companies)
    return sorted_companies

# Task 2 - Print company domains
def print_company_domains():
    for company in company_detail_list:
        print(f"Company {company['name'].split()[-1]}: {company['domain']} ({company['country']})")

# Task 3 - List employees by domain
def get_employees_by_domain():
    result = {}
    # Initialize a dictionary with domains as keys
    for company in company_detail_list:
        result[company['domain']] = []

    # Add employees to the corresponding domain
    for employee in employee_detail_list:
        company = next(c for c in company_detail_list if c['name'] == employee['company'])
        result[company['domain']].append(employee['full_name'])

    print(result)
    return result

# Task 4 - Employees with company country
def get_employees_with_company_country():
    result = []
    # Create a dictionary to map companies to countries
    company_countries = {c['name']: c['country'] for c in company_detail_list}

    for employee in employee_detail_list:
        result.append({
            'full_name': employee['full_name'],
            'company': employee['company'],
            'country': company_countries[employee['company']]
        })
    print(result)
    return result

# Task 5 - Number of employees by nationality
def get_company_employee_nationalities():
    result = []
    for company in company_detail_list:
        nationality_count = {}
        company_employees = [e for e in employee_detail_list if e['company'] == company['name']]

        for employee in company_employees:
            nationality = employee['nationality']
            nationality_count[nationality] = nationality_count.get(nationality, 0) + 1

        result.append({
            'company': company['name'],
            'employee_nationality': nationality_count
        })
    print(result)
    return result

# Execute and display results
print("\nTask 1 - Sorted companies:")
get_sorted_companies()

print("\nTask 2 - Company domains:")
print_company_domains()

print("\nTask 3 - Employees by domain:")
get_employees_by_domain()

print("\nTask 4 - Employees with company country:")
get_employees_with_company_country()

print("\nTask 5 - Number of employees by nationality:")
get_company_employee_nationalities()
```
